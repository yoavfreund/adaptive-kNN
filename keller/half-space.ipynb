```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

from sklearn import datasets

from kdtree import KDTree
```

```python
d = 2 # dimensionality
X = np.random.uniform(-1, 1, size=(500, d))
h = np.random.randn(d) # random half-space vector (via gaussian)
Y = (np.dot(X, h) > 0).astype(np.int)

# visualize
plt.figure(figsize=(5, 5))
X0, X1 = X[Y == 0], X[Y == 1]
plt.scatter(X0[:, 0], X0[:, 1])
plt.scatter(X1[:, 0], X1[:, 1])
plt.show()
```

```python
tree = KDTree(X, Y)
```

```python
fig, ax = plt.subplots()
tree.viz_data()
tree.viz_tree()
plt.show()
```

```python
fig, ax = plt.subplots()
tree.viz_data()
tree.viz_point(np.array([0, 0]), ax)
plt.show()
```
