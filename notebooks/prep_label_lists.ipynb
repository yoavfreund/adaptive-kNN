```python
import mnist

def recast(x):
    return np.array(x,dtype=np.float32)
train_images = recast(mnist.train_images())
train_labels = mnist.train_labels()
test_images = recast(mnist.test_images())
test_labels = mnist.test_labels()
```

```python
m=train_images.shape[0]
def calc_NN(test_image):
    Diff=train_images - test_image
    A=reshape(Diff,[m,-1])
    A=A**2
    S=sum(A**2,axis=1)
    order=argsort(S)
    return order
```

```python
indicesByDistance=[]
for i in arange(test_labels.shape[0]):
    order=np.array(calc_NN(test_images[i]))
    indicesByDistance.append(order)
    labels=train_labels[order]
    label=test_labels[i]
    if i % 10 == 0:
        print(i,label,labels[:10])
```
